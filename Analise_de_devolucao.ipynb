```python
import os
import pandas as pd

lista_arquivo= os.listdir("/content/drive/MyDrive/Curso Básico de Python/Vendas")

tabela_total=pd.DataFrame()
for arquivo in lista_arquivo:
  if "Devolucoes" in arquivo:
    tabela= pd.read_csv(f"/content/drive/MyDrive/Curso Básico de Python/Vendas/{arquivo}")
    tabela_total=tabela_total.append(tabela)

display(tabela_total)
```

**Classificação dos produtos que apresentaram o maior índice de devolução**

```python
tabela_produtos= tabela_total.groupby("Produto").sum()
tabela_produtos= tabela_produtos[["Quantidade Devolvida",]].sort_values(by="Quantidade Devolvida", ascending=False)
display(tabela_produtos)
```

**Classificação dos dispositivos que apresentaram o maior número de prejuizo**

```python
tabela_total["Prejuizo"]=tabela_total["Quantidade Devolvida"]*tabela_total["Preço Unitário"]
tabela_prejuizo=tabela_total.groupby("Produto").sum()
tabela_prejuizo=tabela_prejuizo[["Prejuizo"]].sort_values(by="Prejuizo",ascending=False)
display(tabela_prejuizo)
```

**Classificação das lojas com o maior número de produtos devolvidos**

```python
tabela_cidade=tabela_total.groupby("Loja").sum()
tabela_cidade=tabela_cidade[["Quantidade Devolvida","Prejuizo"]].sort_values(by="Quantidade Devolvida", ascending=False)
display(tabela_cidade)
```

**Análise de produtos com maior número de devolução na capital com menor índice de devolução**

```python
tabela_recife= pd.read_csv("/content/drive/MyDrive/Curso Básico de Python/Vendas/Devolucoes - Recife.csv")

tabela_recife["Prejuizo"]=tabela_recife["Quantidade Devolvida"]*tabela_recife["Preço Unitário"]
tabela_recife_prejuizo=tabela_recife.groupby("Produto").sum()
tabela_recife_prejuizo=tabela_recife_prejuizo[["Quantidade Devolvida","Prejuizo"]].sort_values(by="Prejuizo", ascending=False)
display(tabela_recife_prejuizo)

```
