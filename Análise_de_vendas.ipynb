**Percorrendo todos os arquivos dentro da pasta**
Primeiro deve-se importar a biblioteca os, onde ela foi criada com o intuito de conseguir percorrer arquivos no  dispositivo.

````python
import os
import pandas as pd

lista_arquivo= os.listdir("/content/drive/MyDrive/Curso Básico de Python/Vendas")
````

A lista_arquivo estará armazenando todos os arquivos da pasta de vendas


````python
display(lista_arquivo)
````

Para criar a estrutura de repetição para importar as bases de dados utiliza a condição

````python
tabela_total=pd.DataFrame()
for arquivo in lista_arquivo:
  if"Vendas" in arquivo:
    tabela= pd.read_csv(f"/content/drive/MyDrive/Curso Básico de Python/Vendas/{arquivo}")
    tabela_total=tabela_total.append(tabela)

display(tabela_total)
````

**Análise do produto vendido em quantidade**

-O groupby agrupa a coluna escolhida

-Deve-se falar o que irá ser feito com as outras colunas, no caso, sendo somado pelo .sum()

-Pode selecionar quais colunas que desejam ser vistas

-Pode ordenar utilizando ascending


````python
tabela_produtos= tabela_total.groupby('Produto').sum()
tabela_produtos= tabela_produtos[["Quantidade Vendida"]].sort_values(by="Quantidade Vendida", ascending=True)
display(tabela_produtos)
````

**FATURAMENTO POR DISPOSITIVO**

````python
tabela_total["Faturamento"]=tabela_total["Quantidade Vendida"]*tabela_total["Preco Unitario"]

tabela_faturamento=tabela_total.groupby('Produto').sum()
tabela_faturamento= tabela_faturamento.sort_values(by="Faturamento", ascending=False)
display(tabela_faturamento)
````

**FATURAMENTO POR LOJA**

````python
tabela_loja= tabela_total.groupby('Loja').sum()
tabela_loja= tabela_loja[["Quantidade Vendida","Faturamento"]].sort_values(by="Faturamento", ascending=False)
display(tabela_loja)
````

**Importando biblioteca para a criação de gráficos**

````python
import plotly.express as px

grafico=px.bar(tabela_loja,x=tabela_loja.index,y="Faturamento")
grafico.show()
````
